## Data Understanding:

1. **Nomor Induk Mahasiswa (NIM)** = Nomor Induk Mahasiswa yang digunakan sebagai identifikasi unik alumni.
2. **Tahun Masuk UIN Jakarta** = Tahun alumni mulai menempuh pendidikan di UIN Jakarta.
3. **Tahun Wisuda** = Tahun alumni menyelesaikan sidang dan diwisuda. Bisa diisi dengan tahun sidang jika lebih relevan.
4. **Bulan Wisuda** = Bulan saat alumni diwisuda.
5. **No HP (WhatsApp)** = Nomor telepon alumni.
6. **email** = Email alumni.
7. **Waktu Mendapatkan Pekerjaan** = Lama waktu (dalam bulan) dari saat wisuda hingga alumni mendapatkan pekerjaan pertama.
8. **Status saat ini** = Status terkini alumni, apakah mereka sedang melamar pekerjaan, mengurus rumah tangga, bekerja di sektor BUMN, PNS, swasta, berwirausaha, melanjutkan studi, atau menjadi pengajar.
9. **Tahun Wirausaha** = Tahun alumni memulai usaha, khusus bagi alumni yang menjadi wirausaha.
10. **Bidang wirausaha** = Keterangan singkat tentang bidang usaha yang dijalankan oleh alumni, khusus bagi yang berwirausaha.
11. **Perusahaan** = Nama perusahaan atau instansi tempat alumni bekerja saat ini.
12. **Posisi** = Posisi atau jabatan alumni di tempat kerja mereka saat ini.
13. **Penghasilan** = Perkiraan penghasilan bulanan alumni (take-home pay) dalam juta rupiah.
14. **Univeritas** = Nama universitas dan kota tempat alumni melanjutkan studi, jika ada.
15. **Tahun Masuk Studi Lanjut** = Tahun ketika alumni melanjutkan studi setelah lulus dari UIN Jakarta.
16. **Nama jurusan/program studi setelah Kuliah di FST UIN Jakarta** = Nama jurusan atau program studi yang diambil alumni saat melanjutkan pendidikan.
17. **Kesesuaian Kurikulum** = Skor yang diberikan alumni mengenai kesesuaian kurikulum program studi mereka dengan kebutuhan lapangan pekerjaan. Skala 1-5.
18. **Kesesuaian Bidang Ilmu** = Skor yang diberikan alumni tentang seberapa sesuai bidang ilmu yang mereka pelajari dengan pekerjaan mereka saat ini. Skala 1-5.
19. **Saran** = Komentar, kesan, dan saran alumni untuk pengembangan program studi.
20. **Program Studi** = Program studi yang diambil oleh alumni saat menempuh pendidikan di UIN Jakarta. Ada beberapa pilihan seperti Agribisnis, MAG (Magister), Biologi, Fisika, Kimia, Matematika, SI (Sistem Informasi), TI (Teknik Informatika), Teknik Pertambangan, dan MTI (Magister Teknik Informatika).
21. **Lama Bekerja di Institusi Saat ini** =  Lama waktu alumni bekerja di institusi/perusahaan tempat mereka bekerja saat ini.
22. **Kota tempat bekerja saat Ini** = Kota di mana alumni bekerja saat ini.

```python
import pandas as pd
import numpy as np

# Contoh data, Anda bisa mengganti ini dengan DataFrame Anda
data = {
    'Bidang wirausaha': [
        'ticketing services',
        'NayaOutfit : Usaha di bidang fashion muslimah dengan memenafaatkan e-commerce shopee, tiktok, lazada...',
        'Pendidikan / Training', 'KOS', 'trading agribisnis',
        'Jasa Konsultan',
        'Usaha kuliner, menjual aneka dimsum dengan berbagai toping dijual secara offline maupun online',
        'Membuka les privat bagi siswa SD-SMA di bidang matematika, fisika, kimia, dan menjadi co-founder GeoXplore Indonesia',
        'Explorasi Geofisika khususnya air tanah menggunakan metode elektromagnetik ',
        'Konsultan IT ', 'Bimbingan Belajar', 'Seni dan kerajinan',
        'STEM Education dan Software Development', 'Textile', 'Makanan',
        'Menjual tiket saat ada suatu acara', 'Bisnis souvenir custom CD album (shopee: pinahouse studio)',
        'Jasa traik tunai ', 'Bisnis Olshop', 'tidak ada', 'Kuliner',
        'Digital printing', 'usaha makanan ringan dan kue kering rumahan',
        'makanan ringan dan pernak pernik ', '—',
        'Jasa Wisata Alam dan Jasa Laundry Sepatu', 'Peternakan sapi perah ',
        'Kue', 'Menjual pakaian anak di toko dan online',
        'tahun 2018 saya mendirikan boutique fashion yang bernama lalashopyyy sampai saat ini...',
        'Fashion and beauty'
    ]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Definisikan fungsi untuk mengelompokkan bidang wirausaha
def categorize_business(field):
    if pd.isna(field) or field == '-':
        return 'Tidak Terdefinisi'
    elif 'fashion' in field.lower() or 'pakaian' in field.lower():
        return 'Fashion'
    elif 'kuliner' in field.lower() or 'makanan' in field.lower() or 'dimsum' in field.lower():
        return 'Kuliner'
    elif 'pendidikan' in field.lower() or 'bimbingan' in field.lower() or 'training' in field.lower():
        return 'Pendidikan'
    elif 'konsultan' in field.lower() or 'jasa' in field.lower():
        return 'Jasa'
    elif 'agribisnis' in field.lower() or 'pertanian' in field.lower() or 'peternakan' in field.lower():
        return 'Agribisnis'
    elif 'teknologi' in field.lower() or 'it' in field.lower():
        return 'Teknologi'
    elif 'souvenir' in field.lower() or 'kerajinan' in field.lower():
        return 'Kerajinan'
    else:
        return 'Lainnya'

# Menerapkan fungsi untuk mengelompokkan bidang wirausaha
df['Kategori Wirausaha'] = df['Bidang wirausaha'].apply(categorize_business)

# Menampilkan hasil
print(df[['Bidang wirausaha', 'Kategori Wirausaha']])
```

```python
# Melihat nilai unik dari kolom 'Kategori Wirausaha'
unique_categories = df['Kategori Wirausaha'].unique()

# Menampilkan hasil
print("Kategori Wirausaha yang unik setelah pengelompokan:")
print(unique_categories)
```
